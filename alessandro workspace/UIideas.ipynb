```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets.widgets import FloatSlider, Button, Label, IntSlider, Output
from ipywidgets.widgets import Layout, HBox, VBox
from IPython.display import display
```

```
slider1 = FloatSlider(value=0, min=0, max=100, step = 0.5, description='Level', layout=Layout(width='490px', margin='0 0 5px 0'))
slider2 = IntSlider(value = 10, min = 1, max = 100, step = 1, description="Songs", layout = Layout(width = '490px', margin = '0 0 100px, 0'))
button1 = Button(description='Create Playlist')
button2 = Button(description='Refresh Playlist')
```

```
out = Output(layout={'border': '1px solid black'})
def execute():
    out.clear_output()
    level = slider1.value
    songs = slider2.value
    #here is where find similar songs should go
    print('Level: ', level)
    for i in range(songs):
      print('song'+str(i+1))
  
def refresh():
    out.clear_output()
    print('Refreshing...')

def on_button_clicked1(but):
    from IPython.display import clear_output
    with out:
        clear_output()
    execute()

def on_button_clicked2(but):
    from IPython.display import clear_output
    with out:
      clear_output()
      out.clear_output()
    refresh()

button1.on_click(on_button_clicked1)
button2.on_click(on_button_clicked2)
```
